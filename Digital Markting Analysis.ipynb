    <li><b>Detect outliers:</b> Identify extreme values that may affect model performance.</li>


```python
print("Null values")
df.isnull().sum()
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

# Select only numerical columns for outlier detection
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Boxplots for each numerical feature to visually inspect outliers
plt.figure(figsize=(15, 12))
for i, col in enumerate(num_cols, 1):
    plt.subplot((len(num_cols) // 3) + 1, 3, i)
    sns.boxplot(x=df[col], color='skyblue')
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()


# Z-score method for identifying outliers
from scipy import stats
import numpy as np

z_scores = np.abs(stats.zscore(df[num_cols]))
outliers = (z_scores > 3)

# Count of outliers per column
outlier_counts = pd.DataFrame({
    
    'OutlierCount': (outliers).sum(axis=0)
}).sort_values(by='OutlierCount', ascending=False)

print("Outlier counts per column:")
print(outlier_counts)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('digital_marketing_campaign_dataset.csv')

# Calculate the conversion rate and count of records for each level of PreviousPurchases
bivariate_analysis = df.groupby('PreviousPurchases').agg(
    ConversionRate=('Conversion', 'mean'),
    CustomerCount=('PreviousPurchases', 'size')
).reset_index()

# Calculate the percentage of total customers
total_customers = df.shape[0]
bivariate_analysis['CustomerPercentage'] = (bivariate_analysis['CustomerCount'] / total_customers) * 100

# Format ConversionRate as percentage for better display (for table output)
bivariate_analysis['ConversionRate_Percent'] = (bivariate_analysis['ConversionRate'] * 100).round(2).astype(str) + '%'

# Sort by PreviousPurchases for a logical plot order
bivariate_analysis = bivariate_analysis.sort_values(by='PreviousPurchases', ascending=True)

# --- Visualization ---

# Create the figure and the two axes
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot Conversion Rate on the first axis (ax1) as a line
sns.lineplot(data=bivariate_analysis, x='PreviousPurchases', y='ConversionRate', marker='o', color='tab:blue', ax=ax1, label='Conversion Rate')
ax1.set_xlabel('Number of Previous Purchases')
# Set y-axis to display conversion rate as a percentage for better readability
ax1.set_ylabel('Conversion Rate (Fraction)', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')
ax1.grid(False) # Turn off grid for ax1

# Create a second axis (ax2) that shares the same x-axis
ax2 = ax1.twinx()

# Plot Customer Count on the second axis (ax2) as bars
sns.barplot(data=bivariate_analysis, x='PreviousPurchases', y='CustomerCount', alpha=0.3, color='tab:orange', ax=ax2, label='Customer Count')
ax2.set_ylabel('Customer Count', color='tab:orange')
ax2.tick_params(axis='y', labelcolor='tab:orange')

# Set title and adjust layout
plt.title('Conversion Rate and Customer Count by Previous Purchases')

# Handle legends for both axes
lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
# Place the combined legend outside the plot
ax1.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper left', bbox_to_anchor=(0.0, 1.0))

fig.tight_layout()

# Save the plot
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('digital_marketing_campaign_dataset.csv')

# --- Compute quartile-based bins (without modifying df) ---
min_val = df['LoyaltyPoints'].min()
q1 = df['LoyaltyPoints'].quantile(0.25)
q2 = df['LoyaltyPoints'].quantile(0.50)
q3 = df['LoyaltyPoints'].quantile(0.75)
max_val = df['LoyaltyPoints'].max()

# Define bins and labels (using int for cleaner display)
bins = [min_val - 1, q1, q2, q3, max_val + 1]
labels = [
    f'Low ({int(min_val)}-{int(q1)})',
    f'Mid-Low ({int(q1)+1}-{int(q2)})',
    f'Mid-High ({int(q2)+1}-{int(q3)})',
    f'High ({int(q3)+1}-{int(max_val)})'
]

# Create temporary Series (not added to df)
loyalty_tiers = pd.cut(
    df['LoyaltyPoints'],
    bins=bins,
    labels=labels,
    right=True,
    include_lowest=True
)

# --- Aggregation for visualization only ---
loyalty_conversion_analysis = (
    pd.DataFrame({
        'LoyaltyTiers': loyalty_tiers,
        'Conversion': df['Conversion']
    })
    .groupby('LoyaltyTiers', observed=False)
    .agg(ConversionRate=('Conversion', 'mean'),
         CustomerCount=('LoyaltyTiers', 'size'))
    .reset_index()
)

# --- Visualization ---
user_colors = ['#e6e6fa', '#b19cd9', '#9370db', '#6a0dad']

plt.figure(figsize=(9, 6))
sns.barplot(
    data=loyalty_conversion_analysis,
    x='LoyaltyTiers',
    y='ConversionRate',
    palette=user_colors
)

plt.title('Conversion Rate by Loyalty Point Tier', fontsize=16)
plt.xlabel('Loyalty Point Tier', fontsize=12)
plt.ylabel('Conversion Rate (Fraction)', fontsize=12)
plt.xticks(rotation=15, ha='right')

# Add percentage labels
for index, row in loyalty_conversion_analysis.iterrows():
    plt.text(
        index,
        row['ConversionRate'] + 0.005,
        f'{row["ConversionRate"]:.1%}',
        color='black',
        ha='center',
        fontsize=10
    )

plt.tight_layout()
plt.show()
```

```python
def fit_results(pipe, X_train, X_test, y_train, y_test):

    """ 
    Function fits pipeline on train data and returns model 
    metrics
    """
    # fit pipeline on train data
    pipe.fit(X_train, y_train)
    y_hat_train = pipe.predict(X_train)
    y_hat_test = pipe.predict(X_test)

    # scores
    # Get accuracy for train and test
    train_accuracy = accuracy_score(y_train, y_hat_train)
    test_accuracy = accuracy_score(y_test, y_hat_test)

    print(f"Train Accuracy: {train_accuracy:.2f}, Test Accuracy: {test_accuracy:.2f}")
    print("------------------------------------")

    # Create a confusion matrix
    print("Confusion Matrix")
    cnf_matrix = confusion_matrix(y_test, y_hat_test)
    disp = ConfusionMatrixDisplay(confusion_matrix=cnf_matrix)
    disp.plot()
    plt.show()

    # Get classification report
    report = classification_report(y_test, y_hat_test, target_names=['No', 'Yes'])
    print("Classification Report")
    print("------------------------------------")
    print(report)
```

```python
# -------------------------
# Full pipeline: Preprocessing + Undersample + Oversample + XGBoost + GridSearchCV
# -------------------------
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline as ImbPipeline
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import numpy as np

# -------------------------
# 1. Pipeline
# -------------------------
pipe_xgb = ImbPipeline([
    ('preprocess', preprocess),  # your preprocessing steps
    ('under', RandomUnderSampler(sampling_strategy=0.3, random_state=42)),  # majority reduced to 50%
    ('over', SMOTE(sampling_strategy=0.4, random_state=42)),  # minority increased to 60% of majority
    ('xgboost', XGBClassifier(objective='binary:logistic', random_state=42, scale_pos_weight=2))
])

# -------------------------
# 2. Hyperparameter grid
# -------------------------
param_grid = {
    'xgboost__max_depth': [3, 5, 7],
    'xgboost__learning_rate': [0.1, 0.01, 0.001],
    'xgboost__subsample': [0.5, 0.7, 1]
}

# -------------------------
# 3. GridSearchCV using f1-score (balances precision & recall)
# -------------------------
grid_search = GridSearchCV(pipe_xgb, param_grid=param_grid, cv=5, scoring='f1')
grid_search.fit(X_train, y_train)

print("Best hyperparameters:", grid_search.best_params_)
print("Best CV F1-score:", grid_search.best_score_)

# -------------------------
# 4. Evaluate on test set
# -------------------------
best_model = grid_search.best_estimator_

# Predict probabilities
y_proba = best_model.predict_proba(X_test)[:, 1]

# Tune threshold to improve class 0 precision
```

```python
threshold = 0.53# adjust if needed
y_pred = (y_proba > threshold).astype(int)

# Accuracy
test_accuracy = accuracy_score(y_test, y_pred)
print("Test accuracy:", test_accuracy)

# Confusion Matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_pred, digits=3))

# -------------------------
# 5. Training evaluation (optional)
# -------------------------

```
